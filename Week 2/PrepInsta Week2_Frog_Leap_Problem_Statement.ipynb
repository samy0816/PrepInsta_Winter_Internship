# Problem statement

Create famous 'Frog leap' puzzle game. Try completing the game before starting to get an idea about its working.
[Demonstration](https://www.neok12.com/games/leap-froggies/leap-froggies.htm).


### Rules ###
1. The left set of frogs can only move right, the right set of frogs can only move left.
2. Frogs can move forward one space, or move two spaces by jumping over another frog from opposite side.
3. The puzzle is solved when the two sets of frogs have switched positions.


## Steps to solve the problem:
### Step1:-
- Display green and brown frogs on the left and right sides initially.

Initial Display :-  
```

['G', 'G', 'G', '-', 'B', 'B', 'B']
```
<br>
Here 'G' represents Green frogs on the left side and 'B' represents brown frogs on the right side. The '-' defines the position of empty leaf.
(You can change display according to your imagination or convinience)

### Step2:-
Accept positions of the frog that you want to move.<br>
Example: If we enter position 2 then the game will look like this:-
```
[ 0 ,  1 ,  2 ,  3 ,  4 ,  5 ,  6 ]
['G', 'G', '-', 'G', 'B', 'B', 'B']
```

### Step3:- ###
Define Invalid moves and add conditional 'if' statements accordingly
#### Rules
1. Entered position should be between 0 to 6. Or a character 'q' to quit the game.
2. Entered position cannot be the position of empty leaf.
3. If the selected frog position cannot perform the contraints given in rule 2 then the move is invalid.

### Step4:-
Make the appropriate move by changing the game display.

Now the game should keep running until the player quits, so place all conditional statements inside an infinite loop.<br>

1. We have to `'break'` the loop if the player presses `'q'` and quits.

2. If the move made by player is `'Invalid Move'` then we have to `'continue'` without executing remaining part of the selected iteration.

3. If player wins the game we have to `break` the loop.


```
Infinite loop:
    (inside loop)
    1.Take input
    2.Check all valid and invalid conditions of `pos`.
    3.Make the appropriate move by calculating `pos2`.
    4.Display game
    4.Check winning condition
```

````python
def leap_frog():

  while True:
      no = int(input("Enter the number of frogs (even number): "))

      if no % 2 == 0:
          break
      else:
          print("Wrong Input! Please Enter an Even Number (e.g., 2, 4, 6, 8)")

  divide = no // 2
  i = 0
  j = 0
  positions = []
  num_positions = []

  while j < no + 1:
      positions.append("")
      num_positions.append("")
      j = j + 1

  while i < no + 1:
      if i < divide:
          positions[i] = "G"
          num_positions[i] = i
      elif i == divide:
          positions[i] = "-"
          num_positions[i] = i
      else:
          positions[i] = "B"
          num_positions[i] = i
      i = i + 1

  print(positions)
  Win_Array = positions[::-1]

  while True:
      pos = input("Press q to quit\nEnter position of piece: ")

      if pos == 'q':
          break
      elif pos != 'q':
          pos = int(pos)

      # To check whether the user is entering a valid position
      if pos < 0 or 6 < pos:
          print("Invalid Move")
          continue
      if positions[pos] == '-':
          print("Invalid Move")
          continue

      pos2 = 0

      # If selected frog is Green
      if positions[pos] == 'G':
          if pos + 1 <= 6 and positions[pos + 1] == '-':
              pos2 = pos + 1
          elif pos + 2 <= 6 and positions[pos + 2] == '-' and positions[pos + 1] == 'B':
              pos2 = pos + 2
          else:
              print("Invalid Move")
              continue

      # If selected frog is Brown
      if positions[pos] == 'B':
          if pos - 1 >= 0 and positions[pos - 1] == '-':
              pos2 = pos - 1
          elif pos - 2 >= 0 and positions[pos - 2] == '-' and positions[pos - 1] == 'G':
              pos2 = pos - 2
          else:
              print("Invalid Move")
              continue

      if pos2 >= 0:
          positions[pos], positions[pos2] = positions[pos2], positions[pos]

      moves_left = False
      j = 0
      while j < len(positions):
        if positions[j] == 'G':
            if j + 1 < len(positions) and positions[j + 1] == '-':
                moves_left = True
                break
            elif j + 2 < len(positions) and positions[j + 2] == '-' and positions[j + 1] == 'B':
                moves_left = True
                break
        elif positions[j] == 'B':
            if j - 1 >= 0 and positions[j - 1]
             == '-':
                moves_left = True
                break
            elif j - 2 >= 0 and positions[j - 2] == '-' and positions[j - 1] == 'G':
                moves_left = True
                break
        j = j + 1

      if not moves_left:

            print("The game is stuck!! You Lose :( ")

            choice = input("Do you want to restart the game? (Y/N): ")
            if choice == 'Y':
                      leap_frog()
            elif choice == 'N':
                print("Have a good Day! Goodbye!.")
                break
            else:
                print("Your choice is invalid.Thank you for playing!")
                break

      print(num_positions)
      print(positions)

      if positions == Win_Array:
          print("You Win !!!")
          break

leap_frog()
````

**Logic for customizable no of frogs**  :-
*The user must enter an even number to determine the number of frogs; if the user enters an odd number, the program will print an error and prompt the user to enter the number again.
Next, we use the variable "divide" to determine the number of frogs that must be placed before and after blank leaves by dividing the entered number.
Next, we initiate two lists, one for the locations of the frogs and the other for numerical positions with blank spaces.
Finally, we add Green and Brown Frogs, respectively, while maintaining the middle blank leaf.*


**Logic for Valid Moves**:-  
*In order to determine if any valid moves remain,
   I first made sure that for every frog, the [j+1] or [j+2] position (for the G frog)
   and the [j-1] or [j-2] position (for the B frog) either of them is vacant, allowing the frog to make a move that is allowed.
   The game is stuck if the before-mentioned criteria is not true, as no frog can make a move that is allowable.*
