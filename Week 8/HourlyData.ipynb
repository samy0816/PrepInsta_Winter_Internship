```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

***I have already cleaned dataset by modifying ActivityDay column into hour and date for better analysis***

```python
x1=pd.read_csv("Hourly_Final.csv",index_col=0)
```

```python
x1
```

```python
x1.info(verbose=True)
```

# ***Average Intensity and Total Intensity by ActivityDay(Hourly)***

```python
aggregated_data_steps = x1.groupby('ActivityHour')['AverageIntensity'].sum().reset_index()
fig = px.scatter(aggregated_data_steps, x="ActivityHour", y="AverageIntensity", title='AverageIntensity by ActivityDay')
fig.show()
```

```python
intensity_by_hour = x1.groupby('ActivityDay')['TotalIntensity'].mean().reset_index()
fig = px.line(intensity_by_hour, x='ActivityDay', y='TotalIntensity', title='Total Intensity by Hour')
fig.update_xaxes(title_text='Hour')
fig.update_yaxes(title_text='Total Intensity')
fig.show()
```

# ***Avergae Calories Burned Hourly***

```python
aggregated_data_calories = x1.groupby('ActivityHour')['Calories'].sum().reset_index()
fig = px.scatter(aggregated_data_calories, x="ActivityHour", y="Calories", title='Average Calories burned Hourly')
fig.show()
```
