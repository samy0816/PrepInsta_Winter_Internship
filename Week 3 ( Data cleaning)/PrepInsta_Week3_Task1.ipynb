***3. Outliers :
 Question: Identify potential outliers in numerical indicators (e.g., Age, Salary, Rating). Should outliers be removed or adjusted?***

 -
I have calculated outliers for numerical fields such as age and rating with the help of the IQR method. From the given data, we can see the rating column as some data is missing and some is in the wrong format, so before finding outliers, I first corrected the wrong data and handled missing values by taking the mode of the other values, and to handle the "-1," I had taken the absolute of the column, resulting in converting any value that is in the negative into a positive one (as a rating can't be in the negative).





```python
x2=x["Rating"].mode()[0]
x["Rating"]=x["Rating"].fillna(x2)
x["Rating"]=x["Rating"].abs()

Out=['Age','Rating']
for i in Out:

    Q1 = x[i].quantile(0.25)
    Q3 = x[i].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = x[(x[i] < lower_bound) | (x[i] > upper_bound)]

    print(f"Potential outliers in {i}:")
    print(outliers)
```
