```python
import pandas as pd
x=pd.read_csv("/content/drive/MyDrive/chipotle.tsv",sep="\t")
x
```

***1. Missing Values:Done
Question: Check for missing values in each column (Order ID, Quantity, Item Name, Choice Description, Item Price). How should missing values be handled?***

- As we can see, the sole column containing entries with missing data to be handled is choice_description. We will fill the column with "no description is available for all entries having choice_description as Null" in order to manage the missing data.



```python
x.isnull().sum()
```

```python
x["choice_description"]=x["choice_description"].fillna("No description available")
x
```

***3. Duplicated Entries:
Question: Identify and handle duplicated entries in the dataset. How might duplicates impact analysis, and what is the appropriate action?***

- In order to handle duplicate entries, I had to retain one instance of each entry after looking for entries with the same item name, which indicates multiple items.



```python
dupli=x[x.duplicated(subset=['order_id','item_name'],keep=False)]
x=x.drop_duplicates(subset=['order_id','item_name'],keep='first')
x
```

***4. Quantity and Item Price:
Question: Examine the Quantity and Item Price columns. Are there any inconsistencies or anomalies that need correction?***

- I computed the count of values whose quantity count is the same, as well as the same for item_price in order to identify the number of products with the same prices. I also calculated the item_prices values that are unique.

```python
print("\nQuantity:")
print(x['quantity'].value_counts())


print("\nItem Price:")
print(x['item_price'].value_counts())

print("\nItem Price:")
print(x['item_price'].unique())
```

***5. Choice Description:
Question: Analyze the Choice Description column. How should choices be handled, especially when there are multiple descriptions for a single item?***
- We can concatenate all of the descriptions into a single string that represents the item's description in order to handle the many descriptions.



```python
print("\nchoice_description:")
print(x['choice_description'].value_counts())
```

```python
x['choice_description']= x['choice_description'].str.replace(r'\[|\]', '').str.split(', ').apply(', '.join)
x
```

***6. Handling Special Characters:
Question: Check for special characters in text-based columns (e.g., Item Name, Choice Description). How can these be addressed for consistency?***
- '-' is a special character that appears in certain items in the item_name field. To be consistent, I've substituted " " for the '-'. For choice_descreption, solved it in Question No. 5.


```python
x["item_name"]=x['item_name'].replace('[-]'," ",regex=True)
x
```

***7. Order Id Integrity:
Question: Cross-reference the Order ID column for integrity. Are there any irregularities or patterns that need validation?***

- Certain items share the same order_id, but their item_name differs because we have already eliminated duplicate entries.

***13. Consistent Quantity and Price Units:
Question: Ensure consistency in units for Quantity and Item Price. Should any conversions or adjustments be made for uniform analysis?***

-  I have initially looked to see whether there are any missing data in either of the two columns for the uniform analysis. In addition, I changed the item price column from its original object format to a numerical representation for the uniform analysis.

```python
# To ensure if there are any missing values
quantity_null = x['quantity'].isnull().any()
item_price_null = x['item_price'].isnull().any()

print(quantity_null)
print(item_price_null)

print("\n")

x.info()

print("\n")
# For uniform analysis we can convert Price unit into numeric format
x['item_price']=x['item_price'].replace("[$]"," ",regex=True).str.strip().astype(float)

x.info()
```
