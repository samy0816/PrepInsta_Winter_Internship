# ***Activity day vs Sum of calories burned on that day***

```python
aggregated_data = x1.groupby('ActivityDay')['Calories'].sum().reset_index()
fig = px.scatter(aggregated_data, x='ActivityDay', y='Calories', title='Calories Burned Over Time')

fig.update_layout(xaxis_title='Activity Day', yaxis_title='Calories Burned')

fig.show()
```

# ***Each day sum of activity minutes***

```python


# Assuming 'df' contains your original DataFrame with all activity data

# Aggregate each type of activity minutes separately for each day
aggregated_data = x1.groupby('ActivityDay').agg({
    'SedentaryMinutes': 'sum',
    'ModeratelyActiveDistance': 'sum',
    'LightlyActiveMinutes': 'sum',
    'VeryActiveMinutes': 'sum'
}).reset_index()

# Melt the DataFrame to have a single 'Minutes' column indicating the type of activity
aggregated_data_melted = aggregated_data.melt(id_vars=['ActivityDay'], var_name='ActivityType', value_name='Minutes')

# Plot using Plotly Express
fig = px.line(aggregated_data_melted, x='ActivityDay', y='Minutes', color='ActivityType', title='Activity Minutes Over Time')
fig.update_xaxes(title='Activity Day')
fig.update_yaxes(title='Minutes')
fig.show()
```

# ***Each day average of distance travelled by activity type***

```python
aggregated_data = x1.groupby('ActivityDay').agg({
    'SedentaryActiveDistance': 'mean',
    'LightActiveDistance': 'mean',
    'ModeratelyActiveDistance': 'mean',
    'VeryActiveDistance': 'mean'
}).reset_index()


aggregated_data_melted = aggregated_data.melt(id_vars=['ActivityDay'], var_name='ActivityType', value_name='Distance')


fig = px.line(aggregated_data_melted, x='ActivityDay', y='Distance', color='ActivityType', title='Activity Distance Over Time')
fig.update_xaxes(title='Activity Day')
fig.update_yaxes(title='Distance (in miles)')
fig.show()
```

```python
aggregated_data_steps = x1.groupby('ActivityDay')['StepTotal'].sum().reset_index()
fig=px.line(aggregated_data_steps,x="ActivityDay",y="StepTotal")
fig.show()
```

# ***Segmentation of users into different category depending on Calories and Total Steps***

```python
calories_thresholds = x1['Calories'].quantile([0.25, 0.5, 0.75])
steps_thresholds = x1['StepTotal'].quantile([0.25, 0.5, 0.75])

activity_thresholds = {
    'Very Active': {'calories_min': calories_thresholds[0.75], 'steps_min': steps_thresholds[0.75]},
    'Fairly Active': {'calories_min': calories_thresholds[0.5], 'calories_max': calories_thresholds[0.75],
                      'steps_min': steps_thresholds[0.5], 'steps_max': steps_thresholds[0.75]},
    'Sedentary Active': {'calories_max': calories_thresholds[0.5], 'steps_max': steps_thresholds[0.5]},
    'Lightly Active': {'calories_min': calories_thresholds[0.25], 'steps_min': steps_thresholds[0.25]}
}

def assign_category(row):
    for category, thresholds in activity_thresholds.items():
        if ('calories_min' not in thresholds or row['Calories'] >= thresholds['calories_min']) and \
           ('calories_max' not in thresholds or row['Calories'] <= thresholds['calories_max']) and \
           ('steps_min' not in thresholds or row['StepTotal'] >= thresholds['steps_min']) and \
           ('steps_max' not in thresholds or row['StepTotal'] <= thresholds['steps_max']):
            return category
    return 'Uncategorized'

x1['ActivityCategory'] = x1.apply(assign_category, axis=1)

print(x1[['Id', 'ActivityCategory']])
```
