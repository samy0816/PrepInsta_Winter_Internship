```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
x1=pd.read_csv("Minute_Activity_Final.csv",index_col=0)
```

# ***Data cleaning***

```python
x1
```

```python
x1["ActivityMinute"]=pd.to_datetime(x1["ActivityMinute"])
```

```python
x1.info(verbose=True)
```

```python
x1["ActivityDay"]=x1["ActivityMinute"].dt.date
x1["ActivityHour"]=x1["ActivityMinute"].dt.hour
x1["ActivityMinutes"]=x1["ActivityMinute"].dt.minute
```

```python
x1.drop(columns=['ActivityMinute'], inplace=True)
```

```python
x1
```

```python
x1["ActivityDay"]=pd.to_datetime(x1["ActivityDay"])
```

```python
x1.info(verbose=True)
```

# ***Average Calories Burned Per Minute***

```python
x1['CaloriesPerMinute'] = x1['Calories'] / (x1['ActivityHour'] * 60 + x1['ActivityMinutes'])


average_calories_per_minute = x1.groupby(['ActivityDay', 'ActivityHour', 'ActivityMinutes'])['CaloriesPerMinute'].mean().reset_index()


fig = px.line(average_calories_per_minute, x='ActivityDay', y='CaloriesPerMinute', title='Average Calories Burned per Minute')
fig.show()
```

```python
x1
```

```python
sum_calories_by_day = x1.groupby('IsWeekend')['Calories'].sum().reset_index()
sum_calories_by_day['IsWeekend'] = sum_calories_by_day['IsWeekend'].map({True: 'Weekend', False: 'Weekday'})


barplot_average_calories = px.bar(sum_calories_by_day, x='IsWeekend', y='Calories', title='Sum of Calories Burned on Weekends and Weekdays', labels={'IsWeekend': 'Day Type', 'Calories': 'Average Calories Burned'})
barplot_average_calories.show()
```

```python

average_intensity_by_minute = x1.groupby(['ActivityHour', 'ActivityMinutes'])['Intensity'].mean().reset_index()


lineplot_average_intensity = px.line(average_intensity_by_minute, x='ActivityHour', y='Intensity', title='Average Intensity by Minute', labels={'ActivityHour': 'Hour of Day', 'Intensity': 'Average Intensity'})

lineplot_average_intensity.show()
```
