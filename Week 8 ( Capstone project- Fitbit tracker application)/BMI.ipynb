```python
x1=pd.read_csv("/content/weightLogInfo_merged.csv",index_col=0)
```

# ***Data Cleaning***

```python
x1
```

```python
x1["Date"]=pd.to_datetime(x1["Date"])
```

```python
x1.info(verbose=True)
```

```python
import pandas as pd
import numpy as np

def calculate_height(row):
    if pd.notna(row['WeightKg']) and pd.notna(row['BMI']):
        return np.sqrt(row['WeightKg'] / row['BMI'])
    else:
        return np.nan

x1['Height'] = x1.apply(calculate_height, axis=1)

x1['Height'] *= 3.28084
```

```python
x1['Height']=x1["Height"].round(1)
```

```python
x2=pd.read_csv("/content/Daily_Modified.csv",index_col=0)
```

```python
x3=x1.merge(x2,left_on='Id',right_on='Id')
```

```python
x3
```

```python
x3.info(verbose=True)
```

# ***BMI Distribution and Categorization***

```python
px.histogram(x3, x='BMI', title='BMI Distribution')
```

```python
bmi_categories = {
    'Underweight': (0, 18.5),
    'Normal weight': (18.5, 24.9),
    'Overweight': (25, 29.9),
    'Obese': (30, float('inf'))
}

def get_bmi_category(bmi):
    for category, (lower, upper) in bmi_categories.items():
        if lower <= bmi < upper:
            return category
    return 'Undefined'

x3['BMI Category'] = x3['BMI'].apply(get_bmi_category)
```

```python
x3
```

```python
px.pie(x3, names='BMI Category', title='BMI Category')
```
